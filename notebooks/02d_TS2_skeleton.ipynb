# 02d - TS2 Skeletonization

## Setup

```python
from mcf2swc import *

import logging
logging.basicConfig(level=logging.INFO)

print("✅ Libraries imported successfully!")
```

## Load TS2 Mesh

```python
# Load the TS2 mesh from the data directory
mm = MeshManager(mesh_path="../data/mesh/processed/TS2.obj")
mm.visualize_mesh_3d()
```

## Skeleton

```python
spacing = 10

pls = PolylinesSkeleton(polylines_path="../data/polylines/TS2.polylines.txt")
skel = build_traced_skeleton_graph(mm.mesh, pls, options=TraceOptions(spacing=spacing))
print(skel)
```

```python
skel.plot(node_scale=0.5, backend="plotly");
```

```python
# write swc to file
skel.to_swc(f"../data/swc/TS2_s{spacing}.swc")
```
